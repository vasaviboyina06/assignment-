```python
# Importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Reading the dataset
df = pd.read_csv("D:/diabetes.csv")
```

```python
# Displaying the dataset
df
```

```python
# Describe the dataset
df.describe()
```

```python
# Displaying the correlation of dataset
df.corr()
```

```python
# Plotting a Histogram

# Create a figure with a specified figure size
plt.figure(figsize=(10, 6))

plt.hist(df['BMI'] ,bins=20, color='pink', edgecolor='black')

# Add a title to the plot
plt.title('Histogram of Body Mass Index')

# Adding labels to the plot
plt.xlabel('BMI')
plt.ylabel('Frequency')

# Display
plt.show()
```

```python
# Plotting a Scatter graph

# Create a figure with a specified figure size
plt.figure(figsize=(10, 6))

plt.scatter(df['BloodPressure'],df['Glucose'] ,alpha= 0.75, color='green', edgecolors = 'w')

# Add a title to the plot
plt.title('Relationship Between BloodPressue and Glucose')

# Adding labels to the plot
plt.xlabel('BloodPressue')
plt.ylabel('Glucose')

# plotting the grid
plt.grid(True)

# Display
plt.show()
```

```python
# Plotting a heatmap

# Create a figure with a specified figure size
plt.figure(figsize=(10, 6))

# Selecting the correlation matrix 
correlation_matrix = df.corr()

# Use seaborn function to plot correlation matrix in dataframe
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidth = .5)

# Add a title to the plot
plt.title('Heatmap of Diabetes')

# Display
plt.show()
```
